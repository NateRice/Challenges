## Find Windiest Venue

```python
import pandas as pd
```

```python
games=pd.read_csv('games.csv')
games
```

```python
x = games['wind']
a = games['venue_name']
```

```python
x = x.str.slice(0,2)
y = x.astype(int)
```

```python
#games['wind'] = y
games.insert(14, "wind_int", y)
games
```

```python
windy = games.groupby('venue_name')[['wind_int']].mean()
```

```python
mostWind = windy.sort_values(by=['wind_int'], ascending = False)
```
